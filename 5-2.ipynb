# 5.2 教師あり学習
## 5.2.1 線形回帰学習
$p$個の説明変数を持つ$n$組のデータ$(y_i,x_{i1},\dots,x_{ip}) (i=1,\dots,n)$が与えられたとき、線形回帰モデルは
\begin{equation}
 y_i = b_0 + b_1 x_{i1} + \dots + b_px_{ip} + \varepsilon_i (i=1,\dots,n)
 \tag{5.2.1}
\end{equation}
とあらわされる。ここで、$\varepsilon_i$はそれぞれ独立で、平均は0とする。このモデルは、
\begin{equation*}
 \boldsymbol{y} = 
  \begin{pmatrix}
   y_i \\ \vdots \\ y_n
  \end{pmatrix}
 ,\boldsymbol{X} = 
  \begin{pmatrix}
   1 & x_{11} & \dots & x_{1p} \\
   \vdots & \vdots & \ddots & \vdots \\
   1 & x_{n1} & \dots & x_{np}
  \end{pmatrix}
 ,\boldsymbol{\beta} = 
  \begin{pmatrix}
   b_0 \\ \vdots \\ b_p
  \end{pmatrix}
 ,\boldsymbol{\varepsilon} = 
  \begin{pmatrix}
   \varepsilon_1 \\ \vdots \\ \varepsilon_n
  \end{pmatrix}
\end{equation*}
とすると、
\begin{equation*}
 \boldsymbol{y} = \boldsymbol{X}\boldsymbol{\beta} + \boldsymbol{\varepsilon}
\end{equation*}
と表現できる。この回帰係数$\boldsymbol{\beta}$を最小二乗法を用いて求めるには、
\begin{equation*}
 \sum_{i=1}^n (y_i - b_0 -b_1 x_{i1}-\dots - b_p x_{ip})^2 = (\boldsymbol{y}-\boldsymbol{X}\boldsymbol{\beta})^T(\boldsymbol{y}-\boldsymbol{X}\boldsymbol{\beta})
\end{equation*}
を最小化すればいい。ここで、$\boldsymbol{\beta}$の各成分について微分したものを0とすると、方程式
\begin{equation*}
 \boldsymbol{X}^T\boldsymbol{X}\boldsymbol{\beta}=\boldsymbol{X}^T\boldsymbol{y}
\end{equation*}
が得られる。この方程式を**正規方程式**といい、正規方程式の解が最小二乗法の解となる。ここで、$\boldsymbol{X}$のランクが$p+1$以上であれば$\boldsymbol{X}^T\boldsymbol{X}$は正則となり、解が$\hat{\boldsymbol{\beta}}=(\boldsymbol{X}^T\boldsymbol{X})^{-1}\boldsymbol{X}^T\boldsymbol{y}$と一意に定まる。この推定量$\hat{\boldsymbol{\beta}}$は$\boldsymbol{\beta}$の不偏推定量となる。また、$\varepsilon_i$の分散が一定であるならば、この$\hat{\boldsymbol{\beta}}$は最良線形推定量となる。つまり、$\boldsymbol{\beta}$の一次式を$\boldsymbol{y}$の一次式で推定した場合、$\hat{\boldsymbol{\beta}}$を用いた場合が最も分散が小さくなることが知られている。このことを、**ガウス・マルコフの定理**という。
