# 5.2 教師あり学習
## 5.2.1 線形回帰学習
$p$個の説明変数を持つ$n$組のデータ$(y_i,x_{i1},\dots,x_{ip}) (i=1,\dots,n)$が与えられたとき、線形回帰モデルは
\begin{equation}
 y_i = b_0 + b_1 x_{i1} + \dots + b_px_{ip} + \varepsilon_i (i=1,\dots,n)
 \tag{5.2.1}
\end{equation}
とあらわされる。ここで、$\varepsilon_iはそれぞれ独立で、平均は0とする。このモデルは、
\begin{equation*}
 \boldsymbol{y} = 
  \begin{pmatrix}
   y_i \\ \vdots \\ y_n
  \end{pmatrix}
 ,\boldsymbol{X} = 
  \begin{pmatrix}
   1 & x_{11} & \dots & x_{1p} \\
   \vdots & \vdots & \ddots & \vdots \\
   1 & x_{n1} & \dots & x_{np}
  \end{pmatrix}
 ,\boldsymbol{\beta} = 
  \begin{pmatrix}
   b_0 \\ \vdots \\ b_p
  \end{pmatrix}
 ,\boldsymbol{\varepsilon} = 
  \begin{pmatrix}
   \varepsilon_1 \\ \vdots \\ \varepsilon_n
  \end{pmatrix}
\end{equation*}
とすると、
\begin{equation*}
 \boldsymbol{y} = \boldsymbol{X}\boldsymbol{\beta} + \boldsymbol{\varepsilon}
\end{equation*}
と表現できる。
