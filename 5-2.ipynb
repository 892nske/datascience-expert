## 5.2.1 線形回帰学習
$p$個の説明変数を持つ$n$組のデータ$(y_i,x_{i1},\dots,x_{ip}) (i=1,\dots,n)$が与えられたとき、線形回帰モデルは
\begin{equation}
 y_i = b_0 + b_1 x_{i1} + \dots + b_px_{ip} + \varepsilon_i (i=1,\dots,n)
 \tag{5.2.1}
 \label{eq:linear regression}
\end{equation}
とあらわされる。ここで、$\varepsilon_i$はそれぞれ独立で、平均は0とする。このモデルは、
\begin{equation*}
 \boldsymbol{y} = 
  \begin{pmatrix}
   y_i \\ \vdots \\ y_n
  \end{pmatrix}
 ,\boldsymbol{X} = 
  \begin{pmatrix}
   1 & x_{11} & \dots & x_{1p} \\
   \vdots & \vdots & \ddots & \vdots \\
   1 & x_{n1} & \dots & x_{np}
  \end{pmatrix}
 ,\boldsymbol{\beta} = 
  \begin{pmatrix}
   b_0 \\ \vdots \\ b_p
  \end{pmatrix}
 ,\boldsymbol{\varepsilon} = 
  \begin{pmatrix}
   \varepsilon_1 \\ \vdots \\ \varepsilon_n
  \end{pmatrix}
\end{equation*}
とすると、
\begin{equation*}
 \boldsymbol{y} = \boldsymbol{X}\boldsymbol{\beta} + \boldsymbol{\varepsilon}
\end{equation*}
と表現できる。この回帰係数$\boldsymbol{\beta}$を最小二乗法を用いて求めるには、
\begin{equation*}
 \sum_{i=1}^n (y_i - b_0 -b_1 x_{i1}-\dots - b_p x_{ip})^2 = (\boldsymbol{y}-\boldsymbol{X}\boldsymbol{\beta})^T(\boldsymbol{y}-\boldsymbol{X}\boldsymbol{\beta})
\end{equation*}
を最小化すればいい。ここで、$\boldsymbol{\beta}$の各成分について微分したものを0とすると、方程式
\begin{equation*}
 \boldsymbol{X}^T\boldsymbol{X}\boldsymbol{\beta}=\boldsymbol{X}^T\boldsymbol{y}
\end{equation*}
が得られる。この方程式を**正規方程式**といい、正規方程式の解が最小二乗法の解となる。ここで、$\boldsymbol{X}$のランクが$p+1$以上であれば$\boldsymbol{X}^T\boldsymbol{X}$は正則となり、解が$\hat{\boldsymbol{\beta}}=(\boldsymbol{X}^T\boldsymbol{X})^{-1}\boldsymbol{X}^T\boldsymbol{y}$と一意に定まる。この推定量$\hat{\boldsymbol{\beta}}$は$\boldsymbol{\beta}$の不偏推定量となる。また、$\varepsilon_i$の分散が一定であるならば、この$\hat{\boldsymbol{\beta}}$は最良線形推定量となる。つまり、$\boldsymbol{\beta}$の一次式を$\boldsymbol{y}$の一次式で推定した場合、$\hat{\boldsymbol{\beta}}$を用いた場合が最も分散が小さくなることが知られている。このことを、**ガウス・マルコフの定理**という。

### 自由度調整済み決定係数
次に、この線形回帰モデルの当てはまりの程度について考える。データ$\boldsymbol{y}=(y_1.\dots,y_n)^T$に対する予測値$\hat{\boldsymbol{y}}=(\hat{y_1}.\dots,\hat{y_n})^T=\boldsymbol{X}\hat{\boldsymbol{\beta}}$の当てはまりの程度は**決定係数**
\begin{equation*}
 R^2 = 1 - \frac{\sum_{i=1}^n(y_i-\hat{y_i})^2}{\sum_{i=1}^n(y_i-\overline{y})^2},     \overline{y}=\frac{1}{n}\sum_{i=1}^n y_i
\end{equation*}
で確認できる。決定係数は0から1の値をとり、当てはまりが良いほど決定係数は1に近くなる。つまり、決定係数が小さければ説明変数の線形和が目的変数の変動を十分表せておらず、より目的変数に関連する説明変数を探す必要がある。ここで注意する点として、決定係数は今あるデータに基づいて$\boldsymbol{y}=(y_1.\dots,y_n)^T$に対する$\hat{\boldsymbol{y}}$の当てはまりを評価しているだけで、線形回帰モデル式(5.2.1)が適切かどうかを評価しているわけではない。決定係数の欠点として、目的変数$\boldsymbol{y}$と全く関係ない説明変数を追加しても決定係数が大きくなるという点がある。つまり、決定係数では使っている説明変数のうちどれが不要かということを評価することができない。モデル自体を評価したい場合には、決定係数ではなく**自由度調整済み決定係数**
\begin{equation*}
 R_*^2 = 1 - \frac{\sum_{i=1}^n(y_i-\hat{y_i})^2/(n-p-1)}{\sum_{i=1}^n(y_i-\overline{y})^2/(n-1)} = 1-\frac{n-1}{n-p-1}(1-R^2)
\end{equation*}
が用いられる。$p$は説明変数の個数である。$R_*^2$は$-p/(n-p-1)から1の値をとる。$R_*^2$は$R^2$と異なり、説明変数を追加しても値が大きくなるとは限らず、その値が1に違い程良いモデルであると考えられる。

```python
# ランダムデータを用いて1次元線形回帰を行う
# 説明編巣は2～10の値、目的変数は0～300の値をとる
# y＝ax+bで表される線形回帰モデルを構築する
# データの誤差は値が説明変数の値が大きくなると大きくなるようにする
# 大きな外れ値を人工的にいくつか挿入する
# データセットを作成
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)  # 再現性のためのシード設定
# 説明変数（x）と目的変数（y）の生成
x = np.random.uniform(2, 10, 100)  # 説明変数は2～10の値をとる
y = 30 * x + 20 + np.random.normal(0, 10 * x, 100)  # 目的変数は0～300の値をとる（誤差は説明変数の値が大きくなると大きくなる）
# 大きな外れ値の挿入
outliers_x = np.array([3, 7, 9])
outliers_y = np.array([200, 250, 300])  # 大きな外れ値
x = np.concatenate((x, outliers_x))
y = np.concatenate((y, outliers_y))
# データの可視化
plt.scatter(x, y, color='blue', label='Data Points')
plt.xlabel('x (説明変数)')
plt.ylabel('y (目的変数)')
plt.title('1次元線形回帰のデータセット')
plt.legend()
plt.grid()
```

```python
# 線形回帰モデルの構築
from sklearn.linear_model import LinearRegression
# データを2次元配列に変換
X = x.reshape(-1, 1)
# 線形回帰モデルのインスタンス化と学習
model = LinearRegression()
model.fit(X, y)
# 回帰係数と切片の表示
print(f'回帰係数 (a): {model.coef_[0]:.2f}')
print(f'切片 (b): {model.intercept_:.2f}')
# 回帰直線の描画
x_line = np.linspace(2, 10, 100).reshape(-1, 1)
y_line = model.predict(x_line)
plt.plot(x_line, y_line, color='red', label='Regression Line')
# データと回帰直線の表示
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x_line, y_line, color='red', label='Regression Line')
plt.xlabel('x (説明変数)')
plt.ylabel('y (目的変数)')
plt.title('1次元線形回帰のデータセットと回帰直線')
plt.legend()
plt.grid()
plt.show()
```

```python
# tresiduals vs Fitted valuesのプロット
# 予測値の計算
y_pred = model.predict(X)
# 残差の計算
residuals = y - y_pred
# 残差と予測値のプロット
plt.scatter(y_pred, residuals, color='purple')
plt.axhline(0, color='black', linestyle='--')
plt.xlabel('Fitted values (予測値)')
plt.ylabel('Residuals (残差)')
plt.title('Residuals vs Fitted values')
plt.grid()
plt.show()  
```

```python

# 標準化残差のQQプロット
import scipy.stats as stats
# 標準化残差の計算
standardized_residuals = residuals / np.std(residuals)
# QQプロットの作成
plt.figure(figsize=(6, 6))
stats.probplot(standardized_residuals, dist="norm", plot=plt)
plt.title('QQ Plot of Standardized Residuals')
plt.grid()
plt.show()  
```

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
# import japanize_matplotlib
from statsmodels.nonparametric.smoothers_lowess import lowess

np.random.seed(42)

# -----------------------------
# 1. 写真に近いデータを人工生成
# -----------------------------
n = 95
x = np.random.uniform(2, 10, n)
x.sort()

# x が大きいほどばらつきが大きくなるデータ
noise = np.random.normal(0, 8 + 8 * x, n)
y = 10 + 23 * x + noise

# 高めの点を少し追加
high_idx = x > 7
y[high_idx] += np.random.normal(30, 50, high_idx.sum())

# 強い外れ値を追加
x = np.append(x, 2.4)
y = np.append(y, -250)

# -----------------------------
# 2. 線形回帰
# -----------------------------
X = sm.add_constant(x)
model = sm.OLS(y, X).fit()

fitted = model.fittedvalues
resid = model.resid

influence = model.get_influence()
std_resid = influence.resid_studentized_internal
leverage = influence.hat_matrix_diag
cooks_d = influence.cooks_distance[0]

# -----------------------------
# 3. 図5.1 散布図と回帰直線
# -----------------------------
fig, ax = plt.subplots(figsize=(6, 4))

ax.scatter(x, y, color="black", alpha=0.65, s=25)

x_line = np.linspace(x.min(), x.max(), 100)
X_line = sm.add_constant(x_line)
y_line = model.predict(X_line)

ax.plot(x_line, y_line, color="black", linewidth=1.5, label="regression")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_xlim(1.5, 10.5)
ax.set_ylim(-280, 350)
ax.legend(loc="upper left")
ax.set_title("図5.1　散布図と回帰直線", y=-0.28, fontsize=14)

plt.tight_layout()
plt.show()

# -----------------------------
# 4. 図5.2 回帰診断図
# -----------------------------
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

# Residual vs Fitted
ax = axes[0, 0]
ax.scatter(fitted, resid, color="black", alpha=0.55, s=25)
smooth = lowess(resid, fitted, frac=0.5)
ax.plot(smooth[:, 0], smooth[:, 1], color="black", linewidth=1)
ax.axhline(0, color="gray", linestyle="--", linewidth=1)
ax.set_title("Residual vs. Fitted")
ax.set_xlabel("Fitted values")
ax.set_ylabel("Residuals")

# Normal Q-Q
ax = axes[0, 1]
qq = sm.ProbPlot(std_resid)
theoretical = qq.theoretical_quantiles
ordered = np.sort(std_resid)

ax.scatter(theoretical, ordered, color="black", alpha=0.55, s=25)

q1, q3 = np.percentile(ordered, [25, 75])
tq1, tq3 = np.percentile(theoretical, [25, 75])
slope = (q3 - q1) / (tq3 - tq1)
intercept = q1 - slope * tq1

x_qq = np.array([theoretical.min(), theoretical.max()])
ax.plot(x_qq, intercept + slope * x_qq, color="black", linewidth=1)

ax.set_title("Normal Q-Q")
ax.set_xlabel("Theoretical Quantiles")
ax.set_ylabel("Standardized residuals")

# Scale-Location
ax = axes[1, 0]
sqrt_std = np.sqrt(np.abs(std_resid))
ax.scatter(fitted, sqrt_std, color="black", alpha=0.55, s=25)
smooth = lowess(sqrt_std, fitted, frac=0.5)
ax.plot(smooth[:, 0], smooth[:, 1], color="black", linewidth=1)
ax.set_title("Scale-Location")
ax.set_xlabel("Fitted values")
ax.set_ylabel("√|Standardized residuals|")

# Residuals vs Leverage
ax = axes[1, 1]
ax.scatter(leverage, std_resid, color="black", alpha=0.55, s=25)
smooth = lowess(std_resid, leverage, frac=0.6)
ax.plot(smooth[:, 0], smooth[:, 1], color="black", linewidth=1)
ax.axhline(0, color="gray", linestyle="--", linewidth=1)

# Cook's distance の目安線
p = X.shape[1]  # パラメータ数
h = np.linspace(leverage.min() + 0.001, leverage.max(), 100)

for d in [0.5, 1.0]:
    bound = np.sqrt(d * p * (1 - h) / h)
    ax.plot(h, bound, color="gray", linestyle="--", linewidth=1)
    ax.plot(h, -bound, color="gray", linestyle="--", linewidth=1)

ax.set_title("Residuals vs. Leverage")
ax.set_xlabel("Leverage")
ax.set_ylabel("Standardized residuals")

fig.suptitle("図5.2　回帰診断図", y=0.02, fontsize=14)
plt.tight_layout(rect=[0, 0.05, 1, 1])
plt.show()
```
